```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_and_create_dataframe(url, headers):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table')
    data = []

    for row in table.find_all('tr')[1:11]: 
        columns = row.find_all('td')
        data.append([column.get_text(strip=True) for column in columns])

    
    df = pd.DataFrame(data, columns=['Position', 'Team/Player', 'Matches/Team', 'Points/Team', 'Rating/Player'])

    return df

odi_teams_url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'

headers = {'User-Agent': 'Mozilla/5.0'}

odi_teams_df = scrape_and_create_dataframe(odi_teams_url, headers)

odi_batting_url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting'

odi_batting_df = scrape_and_create_dataframe(odi_batting_url, headers)

odi_bowling_url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling'

odi_bowling_df = scrape_and_create_dataframe(odi_bowling_url, headers)

print("Top 10 ODI Teams in men's Cricket:")
print(odi_teams_df)

print("\nTop 10 men's ODI Batting Players:")
print(odi_batting_df)

print("\nTop 10 men's ODI bowler:")
print(odi_bowling_df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.cnbc.com/World/?region=world"
response = requests.get(url)


soup = BeautifulSoup(response.content, 'html.parser')


articles = soup.find_all('div', class_= 'Card-titleContainer')


headlines = []
times = []
links = []


for article in articles:
    headline = article.find("a").text.strip()
    headlines.append(headline)
    time=article.find("time").text.strip()
    time.append(time)
    link=article.find("a")["href"]
    links.append(link)

data = {
    "Headline": headlines,
    "Time": times,
    "News Link": links
}
df = pd.DataFrame(data)


print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')


articles_container = soup.find("div", class_="pod-listing")


titles = []
authors = []
publication_dates = []
links = []


for article in articles_container.find_all("li"):
    title = article.find('h3').text.strip()
    author = article.find('span', class_='text-xs').text.strip()
    publication_date = article.find('span', class_='text-xs').find_next_sibling("span").text.strip()
    link = article.find('a')["href"]
    
    titles.append(title)
    authors.append(author)
    publication_dates.append(publication_date)
    links.append(link)

data = {
    "Title": titles,
    "Author": authors,
    "Publication Date": publication_dates,
    "Article Link": links
}
df = pd.DataFrame(data)

print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.dineout.co.in/delhi-restaurants"
response = requests.get(url)
response

soup = BeautifulSoup(response.content, 'html.parser')

restaurant_names = soup.find_all('div',class_="restnt-info cursor")
cuisines = soup.find_all('span',class_="double-line-ellipsis")
locations = soup.find_all('div',class_="restnt-loc ellipsis")
ratings = soup.find_all('div',class_="restnt-rating rating-4")
image_urls = soup.find_all('img',class_="no-img"),['data-src']


restaurant_list=[]
cuisine_list=[]
location_list=[]
rating_list=[]
image_url_list=[]


for name in restaurant_names:
    restaurant_list.append(name.text.strip())
for cuisine in cuisines:
    cuisine_list.append(cuisine.text.strip())
for location in locations:    
    location_list.append(location.text.strip())
for rating in ratings:    
    rating_list.append(rating.text.strip())
for image in image_urls:    
    image_url_list.append(image['src'])
    
    
data = {
    "Restaurant Name": restaurant_list,
    "Cuisine": cuisine_list,
    "Location": location_list,
    "Ratings": rating_list,
    "Image URL": image_url_list
}


df = pd.DataFrame(data)

print(df)
```
