```python
driver.get("https://www.bcci.tv")
```

```python
Fixtures=driver.find_element(By.XPATH,'/html/body/header/div[3]/div[1]/ul/div[1]/a[2]')
Fixtures.click()
```

```python
series_name=[]
plays=[]
date=[]
time=[]
```

```python
series_tags=driver.find_elements(By.XPATH,'//h5[@class="match-tournament-name ng-binding"]')
for i in series_tags:
    series=i.text
    series_name.append(series)
```

```python
location_tags=driver.find_elements(By.XPATH,'//span[@class="matchOrderText ng-binding ng-scope t20-tag"]')
for i in location_tags:
    location=i.text
    plays.append(location)
```

```python
date_tags=driver.find_elements(By.XPATH,'//div[@class="match-dates ng-binding"]')
for i in date_tags:
    dates=i.text
    date.append(dates)
```

```python
time_tags=driver.find_elements(By.XPATH,'//div[@class="match-time no-margin ng-binding"]')
for i in time_tags:
    times=i.text
    time.append(times)
```

```python
print(len(series_name),len(plays),len(date),len(time))
```

```python
import pandas as pd
df=pd.DataFrame({'SERIES':series_name,'PLAY':plays,'DATE':date,'TIME':time})
df
```

```python
GDP_of_india=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[2]/ul/li[1]/a')
GDP_of_india.click()
```

```python
Rank=[]
State=[]
GSDP_18_19=[]
GSDP_19_20=[]
Share=[]
GDP=[]
```

```python
rank_tags=driver.find_elements(By.XPATH,'//td[@class="data1"]')
for i in rank_tags:
    rank=i.text
    Rank.append(rank)
```

```python
state_tags=driver.find_elements(By.XPATH,'//td[@class="name"]')
for i in state_tags:
    state=i.text
    State.append(state)
```

```python
GSDP_tags=driver.find_elements(By.XPATH,'//td[@class="data sorting_1"]')
for i in GSDP_tags:
    GSDP=i.text
    GSDP_18_19.append(GSDP)
```

```python
GSDP_19tags=driver.find_elements(By.XPATH,'/html/body/div[3]/div[2]/div[5]/div[1]/div/table/thead/tr[1]/th[3]')
for i in GSDP_19tags:
    GSDP19=i.text
    GSDP_19_20.append(GSDP19)
```

```python
share_tags=driver.find_elements(By.XPATH,'/html/body/div[3]/div[2]/div[5]/div[1]/div/table/tbody/tr[1]/td[5]')
for i in share_tags:
    shares=i.text
    Share.append(shares)
```

```python
gdps_tags=driver.find_elements(By.XPATH,'/html/body/div[3]/div[2]/div[5]/div[1]/div/table/thead/tr[1]/th[5]')
for i in gdps_tags:
    gdps=i.text
    GDP.append(gdps)
```

```python
print(len(Rank),len(State),len(GSDP_18_19),len(GSDP_19_20),len(Share),len(GDP))
```

```python
import pandas as pd
df=pd.DataFrame({'RANK':Rank,'STATE':State,'GSDP18':GSDP_18_19,'GSDP19':GSDP_19_20,'SHARE':Share,'GDPp':GDP})
df
```

```python
driver.get("http://github.com/")
```

```python
open_source=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
open_source.click()
```

```python
Repository_Title=[]
Repository_Description=[]
Contributors_Count=[]
Language_Used=[]
```

```python
rank_tags=driver.find_elements(By.XPATH,'//span[@class="text-normal"]')
for i in rank_tags:
    rank=i.text
    Repository_Title.append(rank)
```

```python
rank_tags=driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
for i in rank_tags:
    rank=i.text
    Repository_Description.append(rank)
    
    try:
        rank_tags=driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
        rank=i.text
        Repository_Description.append(rank)
        
    except NoSuchElementException:
        rank_tags.append('-')
```

```python
rank_tags=driver.find_elements(By.XPATH,'//span[@class="d-inline-block mr-3"]')
for i in rank_tags:
    rank=i.text
    Contributors_Count.append(rank)
```

```python
rank_tags=driver.find_elements(By.XPATH,'//span[@class="repo-language-color"]')
for i in rank_tags:
    rank=i.text
    Language_Used.append(rank)
```

```python
print(len(Repository_Title),len(Repository_Description),len(Contributors_Count),len(Language_Used))
```

```python
import pandas as pd
df=pd.DataFrame({'TITLE':Repository_Title,'DESCRIPTION':Repository_Description,'COUNT':Contributors_Count,'LANGUAGE':Language_Used})
df
```
